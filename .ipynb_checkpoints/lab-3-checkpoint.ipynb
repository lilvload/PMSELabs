**Імпорт бібліотек**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

**Вхідні дані**

```python
x = np.array([0.531, 0.524, 0.541, 0.550, 0.559, 0.620, 0.632, 0.672, 0.682,
            0.689, 0.692, 0.694, 0.698, 0.690, 0.710, 0.720, 0.725, 0.730])
y = np.array([0.620, 0.580, 0.640, 0.650, 0.670, 0.680, 0.695, 0.699, 0.710,
            0.715, 0.725, 0.781, 0.790, 0.795, 0.800, 0.810, 0.850, 0.860])
n = len(x)
```

**1. Знайти точкові незміщені статистичні оцінки b0*, b1*для параметрів b0, b1 парної лінійної функції регресії yi=b0+b1xi. Побудувати на одному малюнку кореляційне поле залежності та графік лінійної регресії.**

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)

b1_star = numerator / denominator
b0_star = y_mean - b1_star * x_mean

print(f"Точкова оцінка b1*: {b1_star:.4f}")
print(f"Точкова оцінка b0*: {b0_star:.4f}")
```

*Побудова графіка*

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Кореляційне поле')
y_predicted = b0_star + b1_star * x
plt.plot(x, y_predicted, color='red', label=f'Лінійна регресія: y = {b0_star:.2f} + {b1_star:.2f}x')
plt.xlabel('Маса дітей (X), кг')
plt.ylabel('Зріст дітей (Y), м')
plt.title('Кореляційне поле та лінійна регресія')
plt.legend()
plt.grid(True)
plt.show()
```

**2. З надійністю y = 0,99 побудувати довірчі інтервали для параметрів b0, b1.**

```python
alpha = 1 - 0.99
degrees_freedom = n - 2
t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)
```

*Оцінка стандартної помилки b1**

```python
y_predicted = b0_star + b1_star * x
residual_sum_squares = np.sum((y - y_predicted)**2)
standard_error_regression = np.sqrt(residual_sum_squares / degrees_freedom)
standard_error_b1 = standard_error_regression / np.sqrt(np.sum((x - x_mean)**2))
```

*Довірчий інтервал для b1*

```python
lower_b1 = b1_star - t_critical * standard_error_b1
upper_b1 = b1_star + t_critical * standard_error_b1
print(f"\nДовірчий інтервал для b1 (99%): ({lower_b1:.4f}, {upper_b1:.4f})")
```

*Оцінка стандартної помилки b0**

```python
standard_error_b0 = standard_error_regression * np.sqrt(1/n + x_mean**2 / np.sum((x - x_mean)**2))
```

*Довірчий інтервал для b0*

```python
lower_b0 = b0_star - t_critical * standard_error_b0
if lower_b0 <= 0:
    lower_b0 = 0

upper_b0 = b0_star + t_critical * standard_error_b0
print(f"Довірчий інтервал для b0 (99%): ({lower_b0:.4f}, {upper_b0:.4f})")
```

**3. При рівні значущості a=0,01 перевірити значущість параметра b1.**

```python
alpha = 0.01
degrees_freedom = n - 2
t_critical_two_tailed = stats.t.ppf(1 - alpha / 2, degrees_freedom)
```

*Розрахунок спостережуваного значення t-статистики для b1*

```python
t_statistic_b1 = b1_star / standard_error_b1

print(f"\nПеревірка значущості параметра b1 при рівні значущості α = {alpha}:")
print(f"Спостережуване значення t-статистики для b1 (t_b1): {t_statistic_b1:.4f}")
print(f"Критичне значення t-розподілу Стьюдента (двосторонній тест, α = {alpha}, df = {degrees_freedom}): ±{t_critical_two_tailed:.4f}")
```

*Порівняння спостережуваного та критичного значень*

```python
if np.abs(t_statistic_b1) > t_critical_two_tailed:
    print(f"|t_b1| ({np.abs(t_statistic_b1):.4f}) > t_критичне ({t_critical_two_tailed:.4f}).")
    print("Отже, нульова гіпотеза про незначимість параметра b1 відхиляється.")
    print("Параметр b1 є статистично значущим на рівні значущості 0.01.")
else:
    print(f"|t_b1| ({np.abs(t_statistic_b1):.4f}) < t_критичне ({t_critical_two_tailed:.4f}).")
    print("Отже, нульова гіпотеза про незначимість параметра b1 приймається.")
    print("Немає достатньо доказів, щоб стверджувати статистичну значущість параметра b1 на рівні значущості 0.01.")
```

**4. З надійністю y= 0,99 побудувати довірчий інтервал для функції регресії yi=b0+b1xi та зобразити його графічно.**

```python
confidence_level_regression = 0.99
alpha_regression = 1 - confidence_level_regression
t_critical_regression = stats.t.ppf(1 - alpha_regression / 2, degrees_freedom)
```

*Розрахунок стандартної помилки прогнозованого значення регресії*

```python
standard_error_y_hat = standard_error_regression * np.sqrt(1/n + (x - x_mean)**2 / np.sum((x - x_mean)**2))
```

*Розрахунок меж довірчого інтервалу*

```python
lower_bound_regression = y_predicted - t_critical_regression * standard_error_y_hat
upper_bound_regression = y_predicted + t_critical_regression * standard_error_y_hat
```

*Візуалізація довірчого інтервалу для регресії*

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Кореляційне поле')
plt.plot(x, y_predicted, color='red', label=f'Лінійна регресія: y = {b0_star:.2f} + {b1_star:.2f}x')
plt.plot(x, lower_bound_regression, color='green', linestyle='--', label=f'{confidence_level_regression*100}% довірчий інтервал')
plt.plot(x, upper_bound_regression, color='green', linestyle='--')
plt.xlabel('Маса дітей (X), кг')
plt.ylabel('Зріст дітей (Y), м')
plt.title(f'Лінійна регресія з {confidence_level_regression*100}% довірчим інтервалом для функції регресії')
plt.legend()
plt.grid(True)
```

**5. Обчислити парний коефіцієнт кореляції rxy.**

```python
numerator_r = np.sum((x - x_mean) * (y - y_mean))
denominator_r = np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))
r_xy = numerator_r / denominator_r
print(f"\nПарний коефіцієнт кореляції r_xy: {r_xy:.4f}")
```

```python
numerator_r = np.sum((x - x_mean) * (y - y_mean))
denominator_r = np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))

r_xy = numerator_r / denominator_r
print(f"\nПарний коефіцієнт кореляції r_xy: {r_xy:.4f}")
```

**6. З надійністю y = 0,99 побудувати довірчий інтервал для прогнозованих індивідуальних значень та зобразити його графічно.**

```python
confidence_level_prediction = 0.99
alpha_prediction = 1 - confidence_level_prediction
t_critical_prediction = stats.t.ppf(1 - alpha_prediction / 2, degrees_freedom)
```

*Розрахунок стандартної помилки прогнозування індивідуального значення*

```python
standard_error_prediction = standard_error_regression * np.sqrt(1 + 1/n + (x - x_mean)**2 / np.sum((x - x_mean)**2))
```

*Розрахунок меж довірчого інтервалу для прогнозування*

```python
lower_bound_prediction = y_predicted - t_critical_prediction * standard_error_prediction
upper_bound_prediction = y_predicted + t_critical_prediction * standard_error_prediction
```

*Візуалізація довірчого інтервалу для прогнозування*

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Спостережувані значення')
plt.plot(x, y_predicted, color='red', label=f'Лінійна регресія: y = {b0_star:.2f} + {b1_star:.2f}x')
plt.plot(x, lower_bound_prediction, color='orange', linestyle='--', label=f'{confidence_level_prediction*100}% довірчий інтервал прогнозу')
plt.plot(x, upper_bound_prediction, color='orange', linestyle='--')
plt.xlabel('Маса дітей (X), кг')
plt.ylabel('Зріст дітей (Y), м')
plt.title(f'Лінійна регресія з {confidence_level_prediction*100}% довірчим інтервалом для прогнозованих значень')
plt.legend()
plt.grid(True)
plt.show()
```
